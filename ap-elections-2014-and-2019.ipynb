Percentage of Missing values


```python
Total = data_2014.isnull().sum().sort_values(ascending=False)   
```

```python
Percent = (data_2014.isnull().sum()*100/data_2014.isnull().count()).sort_values(ascending=False)   
```

```python
missing_data_2014 = pd.concat([Total, Percent], axis = 1, keys = ['Total', 'Percentage of Missing Values'])    
missing_data_2014.style.background_gradient(cmap="Oranges")
```
